```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh data
data = {
    'Pendapatan': [5000, 7000, 8000, 12000, 15000],
    'Pengeluaran': [2000, 3000, 4000, 7000, 9000],
    'Tabungan': [3000, 4000, 4000, 5000, 6000]
}
df = pd.DataFrame(data)

# Menghitung korelasi
corr_matrix = df.corr()

# Menampilkan heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Generate Data
df = pd.DataFrame(np.random.rand(10, 5) * 10, columns=['F1', 'F2', 'F3', 'F4', 'F5'])

# PCA dengan 2 komponen
df_pca = pd.DataFrame(PCA(n_components=2).fit_transform(StandardScaler().fit_transform(df)), columns=['PC1', 'PC2'])

# Print dan Plot
print(df_pca)
plt.scatter(df_pca['PC1'], df_pca['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Result')
plt.grid()
plt.show()
```
